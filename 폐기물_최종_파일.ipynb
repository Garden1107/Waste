# 폐기물 처리

```python
import pandas as pd
process=pd.read_csv('/content/폐기물_처리현황_사업장배출시설계폐기물_20231128224901.csv'
,encoding='cp949',nrows=33,skiprows=[0,1])
process
```

```python
process=process.drop(index=[0,1,15])
```

```python
process['일반처리방식']=process['매립']+process['소각']
process
```

```python
process['재활용률']=process['재활용']/process['일반처리방식']
process
```

```python
process_sorted=process.sort_values('재활용률',ascending=True)
process_sorted
```

```python
pro_result=process_sorted.head(10)
pro_result
```

```python
pro_result['최종구분']=pro_result['구분(2)']+pro_result['구분(3)']+pro_result['구분(4)'] #폐기물 종류의 구분을 위해서 이 코드르르 ㅅ썼습니다..ㅏ...
pro_result
```

```python
!pip install koreanize-matplotlib
import koreanize_matplotlib
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(pro_result['최종구분'], pro_result['재활용률'],
        color='skyblue',label='재활용률')
plt.title('소재별 재활용률')
plt.legend()
plt.xlabel('소재')
plt.ylabel('재활용률(재활용/일반처리방식)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 지역별 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
!pip install koreanize-matplotlib
import koreanize_matplotlib
```

```python
data=pd.read_csv('/content/폐기물_처리현황_사업장배출시설계폐기물_20231128224901.csv'
,encoding='cp949',skiprows=[0,1])
data
```

```python
data['폐기물 종류']=data['구분(2)']+data['구분(3)']+data['구분(4)']
data
```

```python
data = data.drop(data[data['폐기물 종류'] == '총계소계소계'].index)
data
```

```python
data = data.drop(data[data['구분(3)'] == '소계'].index)
data
```

```python


def make_graph():
  area = input("폐기물 재활용 현황을 알고 싶은 지역명을 입력해주세요: ")
  valid_areas = ['서울','부산','대구','인천','광주','대전','울산','세종','경기'
  ,'강원','충북','충남','전북','전남','경북','경남','제주']

  if area in valid_areas and len(area) == 2:
      print("올바른 지역명을 입력했습니다.")
      choice = data[data['구분(1)'] == area]

      plt.figure(figsize=(100, 50))

      plt.bar(choice['폐기물 종류'],choice['재활용'],label=area + " 재활용률")

      plt.title(area + " 재활용량",fontsize=100)
      plt.xlabel('소재',fontsize=80)
      plt.ylabel(area+"재활용량(ton)",fontsize=80)
      plt.xticks(rotation=45, ha='right',fontsize=50)
      plt.yticks(fontsize=50)

      plt.legend(fontsize=80)
      plt.show()

  else:
      print("지역명이 틀렸거나 두 글자의 시/도명이 아닙니다.")
      make_graph()


make_graph()
```
